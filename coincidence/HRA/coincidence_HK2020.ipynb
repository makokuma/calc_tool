```python
import pandas as pd
import numpy as np

# =========================
# inputs
# =========================
csvfile_1 = '/mnt/jet12/makoto/extract_senjo/ext_sun_re/csv/100-80/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_100-80_newarea_True.csv'
THS = '090-70'
csvfile_2 = f'/mnt/jet12/makoto/extract_senjo/condition_2_re/csv_090-70_25/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_{THS}_newarea_True.csv'

df1 = pd.read_csv(csvfile_1, dtype={'dtst': str, 'dten': str})
df2 = pd.read_csv(csvfile_2, dtype={'dtst': str, 'dten': str})

# =========================
# parameters (A/B/C)
# =========================
# A: time conditions
TIME_MARGIN_H = 12          # dtst-12h ~ dten+12h
MIN_OVERLAP_H = 3           # 重なり最小時間（誤検出抑制なら3〜6hがおすすめ）
MAX_CENTER_DIFF_H = 12      # 中心時刻差の上限

# B: position condition (radius in "grid" unit)
POS_RADIUS_GRID = 0.3       # 半径（円）で 0.3 grid

# C: scoring
ALPHA_H = 0.2               # 時間差(時間)の重み（距離と同じ尺度にしたい時に調整）
# score = dist_grid + ALPHA_H * center_diff_hours

# datetime format (必要なら指定)
#DT_FMT = None
# 例:
DT_FMT = "%Y%m%d%H"           # 2020070403 みたいなやつ
# DT_FMT = "%Y-%m-%d %H:%M:%S"  # 2020-07-04 03:00:00
# DT_FMT を None のままなら推定

# =========================
# helpers
# =========================
def parse_dt(s: pd.Series) -> pd.Series:
    if DT_FMT is None:
        return pd.to_datetime(s, errors="coerce")
    else:
        return pd.to_datetime(s, format=DT_FMT, errors="coerce")

def overlap_hours(a_start, a_end, b_start, b_end) -> float:
    # returns overlap duration in hours (>=0)
    latest_start = max(a_start, b_start)
    earliest_end = min(a_end, b_end)
    if earliest_end <= latest_start:
        return 0.0
    return (earliest_end - latest_start).total_seconds() / 3600.0

# =========================
# preprocess: parse times
# =========================
for d in (df1, df2):
    d["dtst_dt"] = parse_dt(d["dtst"])
    d["dten_dt"] = parse_dt(d["dten"])

# もし NaT が多いと一致判定できないので、ここで確認用（printしたくなければ消してOK）
print("df1 dt parse NaT:", df1["dtst_dt"].isna().sum(), df1["dten_dt"].isna().sum())
print("df2 dt parse NaT:", df2["dtst_dt"].isna().sum(), df2["dten_dt"].isna().sum())

# center time
df1["tcenter"] = df1["dtst_dt"] + (df1["dten_dt"] - df1["dtst_dt"]) / 2
df2["tcenter"] = df2["dtst_dt"] + (df2["dten_dt"] - df2["dtst_dt"]) / 2

# ensure lat/lon numeric (完全一致をやめて距離判定なのでfloatに)
df1["lat"] = pd.to_numeric(df1["lat"], errors="coerce")
df1["lon"] = pd.to_numeric(df1["lon"], errors="coerce")
df2["lat"] = pd.to_numeric(df2["lat"], errors="coerce")
df2["lon"] = pd.to_numeric(df2["lon"], errors="coerce")

# =========================
# main matching (list all candidates)
# =========================
time_margin = pd.Timedelta(hours=TIME_MARGIN_H)

candidates_rows = []   # 全候補（リスト出力用）
best_rows = []         # ベスト一致
notfound_rows = []     # 未一致（df1）

# df2 を時刻で絞るために、dtst/dtenの列だけ先に持っておく（高速化の最低限）
df2_dtst = df2["dtst_dt"].values
df2_dten = df2["dten_dt"].values
df2_tcen = df2["tcenter"].values
df2_lat = df2["lat"].values
df2_lon = df2["lon"].values

for i in range(len(df1)):
    r1 = df1.iloc[i]

    # 欠損があると判定不能なのでスキップ扱い（必要ならログへ）
    if pd.isna(r1["dtst_dt"]) or pd.isna(r1["dten_dt"]) or pd.isna(r1["lat"]) or pd.isna(r1["lon"]):
        notfound_rows.append({
            "df1_i": i,
            "reason": "df1 missing dt/lat/lon",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A: 時間の粗い候補抽出（±12hの範囲に“重なり得る”もの）
    a_start = r1["dtst_dt"] - time_margin
    a_end   = r1["dten_dt"] + time_margin

    # df2イベント区間 [dtst_dt, dten_dt] が [a_start, a_end] と重なり得る条件
    # b_start <= a_end かつ b_end >= a_start
    time_mask = (df2_dtst <= a_end) & (df2_dten >= a_start)

    idx_time = np.where(time_mask)[0]
    if idx_time.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "no time candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- B: 位置（円距離）で絞る
    dlat = df2_lat[idx_time] - float(r1["lat"])
    dlon = df2_lon[idx_time] - float(r1["lon"])
    dist_grid = np.sqrt(dlat*dlat + dlon*dlon)

    pos_mask = dist_grid <= POS_RADIUS_GRID
    idx_tp = idx_time[pos_mask]
    dist_tp = dist_grid[pos_mask]

    if idx_tp.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time ok but no position candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A(強化): 重なり量 + 中心時刻差で絞る
    # 重なりは「元のイベント区間」同士で評価（マージンは候補抽出だけに使う）
    keep = []
    for k, j in enumerate(idx_tp):
        ov_h = overlap_hours(r1["dtst_dt"], r1["dten_dt"], df2.iloc[j]["dtst_dt"], df2.iloc[j]["dten_dt"])
        center_diff_h = abs((r1["tcenter"] - df2_tcen[j]).total_seconds()) / 3600.0

        if (ov_h >= MIN_OVERLAP_H) and (center_diff_h <= MAX_CENTER_DIFF_H):
            keep.append((j, dist_tp[k], ov_h, center_diff_h))

    if len(keep) == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time&pos ok but overlap/center fail",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
            "df1_hrid": r1["hrid"],
        })
        continue

    # ---- C: スコアで最良一致を選ぶ（ただし候補リストは全部残す）
    # score = dist_grid + ALPHA_H * center_diff_h
    cand_list = []
    for (j, distg, ov_h, cd_h) in keep:
        score = distg + ALPHA_H * cd_h
        cand_list.append((j, distg, ov_h, cd_h, score))
        candidates_rows.append({
            "df1_i": i,
            "df2_i": int(j),
            "df1_hrid": r1["hrid"],
            "df2_hrid": df2.iloc[j]["hrid"],
            "dist_grid": float(distg),
            "overlap_h": float(ov_h),
            "center_diff_h": float(cd_h),
            "score": float(score),

            # ぱっと確認用：最小限のメタも付与（必要に応じて増やしてOK）
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": float(r1["lat"]),
            "df1_lon": float(r1["lon"]),
            "df2_dtst": df2.iloc[j]["dtst"],
            "df2_dten": df2.iloc[j]["dten"],
            "df2_lat": float(df2.iloc[j]["lat"]),
            "df2_lon": float(df2.iloc[j]["lon"]),
        })

    # best = min score
    cand_list.sort(key=lambda x: x[4])
    best_j, best_distg, best_ov, best_cd, best_score = cand_list[0]

    best_rows.append({
        "df1_i": i,
        "df2_i": int(best_j),
        "df1_hrid": r1["hrid"],
        "df2_hrid": df2.iloc[j]["hrid"],
        "dist_grid": float(best_distg),
        "overlap_h": float(best_ov),
        "center_diff_h": float(best_cd),
        "score": float(best_score),
        "n_candidates": len(cand_list),

        "df1_dtst": r1["dtst"],
        "df1_dten": r1["dten"],
        "df1_lat": float(r1["lat"]),
        "df1_lon": float(r1["lon"]),
        "df2_dtst": df2.iloc[best_j]["dtst"],
        "df2_dten": df2.iloc[best_j]["dten"],
        "df2_lat": float(df2.iloc[best_j]["lat"]),
        "df2_lon": float(df2.iloc[best_j]["lon"]),
    })

# =========================
# outputs
# =========================
cand_df = pd.DataFrame(candidates_rows).sort_values(["df1_i", "score"])
best_df = pd.DataFrame(best_rows).sort_values(["df1_i"])
nf_df   = pd.DataFrame(notfound_rows).sort_values(["df1_i"])

cand_csv = f"match_candidates_{THS}_A+B+C.csv"
best_csv = f"match_best_{THS}_A+B+C.csv"
nf_csv   = f"match_notfound_{THS}_A+B+C.csv"

cand_df.to_csv(cand_csv, index=False)
best_df.to_csv(best_csv, index=False)
nf_df.to_csv(nf_csv, index=False)

print("saved:", cand_csv)
print("saved:", best_csv)
print("saved:", nf_csv)
print("counts: candidates=", len(cand_df), "best=", len(best_df), "notfound=", len(nf_df))
```

```python
import pandas as pd
import numpy as np

# =========================
# inputs
# =========================
csvfile_1 = '/mnt/jet12/makoto/extract_senjo/ext_sun_re/csv/100-80/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_100-80_newarea_True.csv'
THS = '080-70'
csvfile_2 = f'/mnt/jet12/makoto/extract_senjo/condition_3_re/csv_080-70/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_{THS}_newarea_True.csv'

df1 = pd.read_csv(csvfile_1, dtype={'dtst': str, 'dten': str})
df2 = pd.read_csv(csvfile_2, dtype={'dtst': str, 'dten': str})

# =========================
# parameters (A/B/C)
# =========================
# A: time conditions
TIME_MARGIN_H = 12          # dtst-12h ~ dten+12h
MIN_OVERLAP_H = 3           # 重なり最小時間（誤検出抑制なら3〜6hがおすすめ）
MAX_CENTER_DIFF_H = 12      # 中心時刻差の上限

# B: position condition (radius in "grid" unit)
POS_RADIUS_GRID = 0.3       # 半径（円）で 0.3 grid

# C: scoring
ALPHA_H = 0.2               # 時間差(時間)の重み（距離と同じ尺度にしたい時に調整）
# score = dist_grid + ALPHA_H * center_diff_hours

# datetime format (必要なら指定)
#DT_FMT = None
# 例:
DT_FMT = "%Y%m%d%H"           # 2020070403 みたいなやつ
# DT_FMT = "%Y-%m-%d %H:%M:%S"  # 2020-07-04 03:00:00
# DT_FMT を None のままなら推定

# =========================
# helpers
# =========================
def parse_dt(s: pd.Series) -> pd.Series:
    if DT_FMT is None:
        return pd.to_datetime(s, errors="coerce")
    else:
        return pd.to_datetime(s, format=DT_FMT, errors="coerce")

def overlap_hours(a_start, a_end, b_start, b_end) -> float:
    # returns overlap duration in hours (>=0)
    latest_start = max(a_start, b_start)
    earliest_end = min(a_end, b_end)
    if earliest_end <= latest_start:
        return 0.0
    return (earliest_end - latest_start).total_seconds() / 3600.0

# =========================
# preprocess: parse times
# =========================
for d in (df1, df2):
    d["dtst_dt"] = parse_dt(d["dtst"])
    d["dten_dt"] = parse_dt(d["dten"])

# もし NaT が多いと一致判定できないので、ここで確認用（printしたくなければ消してOK）
print("df1 dt parse NaT:", df1["dtst_dt"].isna().sum(), df1["dten_dt"].isna().sum())
print("df2 dt parse NaT:", df2["dtst_dt"].isna().sum(), df2["dten_dt"].isna().sum())

# center time
df1["tcenter"] = df1["dtst_dt"] + (df1["dten_dt"] - df1["dtst_dt"]) / 2
df2["tcenter"] = df2["dtst_dt"] + (df2["dten_dt"] - df2["dtst_dt"]) / 2

# ensure lat/lon numeric (完全一致をやめて距離判定なのでfloatに)
df1["lat"] = pd.to_numeric(df1["lat"], errors="coerce")
df1["lon"] = pd.to_numeric(df1["lon"], errors="coerce")
df2["lat"] = pd.to_numeric(df2["lat"], errors="coerce")
df2["lon"] = pd.to_numeric(df2["lon"], errors="coerce")

# =========================
# main matching (list all candidates)
# =========================
time_margin = pd.Timedelta(hours=TIME_MARGIN_H)

candidates_rows = []   # 全候補（リスト出力用）
best_rows = []         # ベスト一致
notfound_rows = []     # 未一致（df1）

# df2 を時刻で絞るために、dtst/dtenの列だけ先に持っておく（高速化の最低限）
df2_dtst = df2["dtst_dt"].values
df2_dten = df2["dten_dt"].values
df2_tcen = df2["tcenter"].values
df2_lat = df2["lat"].values
df2_lon = df2["lon"].values

for i in range(len(df1)):
    r1 = df1.iloc[i]

    # 欠損があると判定不能なのでスキップ扱い（必要ならログへ）
    if pd.isna(r1["dtst_dt"]) or pd.isna(r1["dten_dt"]) or pd.isna(r1["lat"]) or pd.isna(r1["lon"]):
        notfound_rows.append({
            "df1_i": i,
            "reason": "df1 missing dt/lat/lon",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A: 時間の粗い候補抽出（±12hの範囲に“重なり得る”もの）
    a_start = r1["dtst_dt"] - time_margin
    a_end   = r1["dten_dt"] + time_margin

    # df2イベント区間 [dtst_dt, dten_dt] が [a_start, a_end] と重なり得る条件
    # b_start <= a_end かつ b_end >= a_start
    time_mask = (df2_dtst <= a_end) & (df2_dten >= a_start)

    idx_time = np.where(time_mask)[0]
    if idx_time.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "no time candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- B: 位置（円距離）で絞る
    dlat = df2_lat[idx_time] - float(r1["lat"])
    dlon = df2_lon[idx_time] - float(r1["lon"])
    dist_grid = np.sqrt(dlat*dlat + dlon*dlon)

    pos_mask = dist_grid <= POS_RADIUS_GRID
    idx_tp = idx_time[pos_mask]
    dist_tp = dist_grid[pos_mask]

    if idx_tp.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time ok but no position candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A(強化): 重なり量 + 中心時刻差で絞る
    # 重なりは「元のイベント区間」同士で評価（マージンは候補抽出だけに使う）
    keep = []
    for k, j in enumerate(idx_tp):
        ov_h = overlap_hours(r1["dtst_dt"], r1["dten_dt"], df2.iloc[j]["dtst_dt"], df2.iloc[j]["dten_dt"])
        center_diff_h = abs((r1["tcenter"] - df2_tcen[j]).total_seconds()) / 3600.0

        if (ov_h >= MIN_OVERLAP_H) and (center_diff_h <= MAX_CENTER_DIFF_H):
            keep.append((j, dist_tp[k], ov_h, center_diff_h))

    if len(keep) == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time&pos ok but overlap/center fail",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
            "df1_hrid": r1["hrid"],
        })
        continue

    # ---- C: スコアで最良一致を選ぶ（ただし候補リストは全部残す）
    # score = dist_grid + ALPHA_H * center_diff_h
    cand_list = []
    for (j, distg, ov_h, cd_h) in keep:
        score = distg + ALPHA_H * cd_h
        cand_list.append((j, distg, ov_h, cd_h, score))
        candidates_rows.append({
            "df1_i": i,
            "df2_i": int(j),
            "df1_hrid": r1["hrid"],
            "df2_hrid": df2.iloc[j]["hrid"],
            "dist_grid": float(distg),
            "overlap_h": float(ov_h),
            "center_diff_h": float(cd_h),
            "score": float(score),

            # ぱっと確認用：最小限のメタも付与（必要に応じて増やしてOK）
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": float(r1["lat"]),
            "df1_lon": float(r1["lon"]),
            "df2_dtst": df2.iloc[j]["dtst"],
            "df2_dten": df2.iloc[j]["dten"],
            "df2_lat": float(df2.iloc[j]["lat"]),
            "df2_lon": float(df2.iloc[j]["lon"]),
        })

    # best = min score
    cand_list.sort(key=lambda x: x[4])
    best_j, best_distg, best_ov, best_cd, best_score = cand_list[0]

    best_rows.append({
        "df1_i": i,
        "df2_i": int(best_j),
        "df1_hrid": r1["hrid"],
        "df2_hrid": df2.iloc[j]["hrid"],
        "dist_grid": float(best_distg),
        "overlap_h": float(best_ov),
        "center_diff_h": float(best_cd),
        "score": float(best_score),
        "n_candidates": len(cand_list),

        "df1_dtst": r1["dtst"],
        "df1_dten": r1["dten"],
        "df1_lat": float(r1["lat"]),
        "df1_lon": float(r1["lon"]),
        "df2_dtst": df2.iloc[best_j]["dtst"],
        "df2_dten": df2.iloc[best_j]["dten"],
        "df2_lat": float(df2.iloc[best_j]["lat"]),
        "df2_lon": float(df2.iloc[best_j]["lon"]),
    })

# =========================
# outputs
# =========================
cand_df = pd.DataFrame(candidates_rows).sort_values(["df1_i", "score"])
best_df = pd.DataFrame(best_rows).sort_values(["df1_i"])
nf_df   = pd.DataFrame(notfound_rows).sort_values(["df1_i"])

cand_csv = f"match_candidates_{THS}_A+B+C.csv"
best_csv = f"match_best_{THS}_A+B+C.csv"
nf_csv   = f"match_notfound_{THS}_A+B+C.csv"

cand_df.to_csv(cand_csv, index=False)
best_df.to_csv(best_csv, index=False)
nf_df.to_csv(nf_csv, index=False)

print("saved:", cand_csv)
print("saved:", best_csv)
print("saved:", nf_csv)
print("counts: candidates=", len(cand_df), "best=", len(best_df), "notfound=", len(nf_df))
```

```python
import pandas as pd
import numpy as np

# =========================
# inputs
# =========================
csvfile_1 = '/mnt/jet12/makoto/extract_senjo/ext_sun_re/csv/100-80/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_100-80_newarea_True.csv'
THS = '070-50'
csvfile_2 = f'/mnt/jet12/makoto/extract_senjo/condition_4/csv/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_{THS}_newarea_True.csv'

df1 = pd.read_csv(csvfile_1, dtype={'dtst': str, 'dten': str})
df2 = pd.read_csv(csvfile_2, dtype={'dtst': str, 'dten': str})

# =========================
# parameters (A/B/C)
# =========================
# A: time conditions
TIME_MARGIN_H = 12          # dtst-12h ~ dten+12h
MIN_OVERLAP_H = 3           # 重なり最小時間（誤検出抑制なら3〜6hがおすすめ）
MAX_CENTER_DIFF_H = 12      # 中心時刻差の上限

# B: position condition (radius in "grid" unit)
POS_RADIUS_GRID = 0.3       # 半径（円）で 0.3 grid

# C: scoring
ALPHA_H = 0.2               # 時間差(時間)の重み（距離と同じ尺度にしたい時に調整）
# score = dist_grid + ALPHA_H * center_diff_hours

# datetime format (必要なら指定)
#DT_FMT = None
# 例:
DT_FMT = "%Y%m%d%H"           # 2020070403 みたいなやつ
# DT_FMT = "%Y-%m-%d %H:%M:%S"  # 2020-07-04 03:00:00
# DT_FMT を None のままなら推定

# =========================
# helpers
# =========================
def parse_dt(s: pd.Series) -> pd.Series:
    if DT_FMT is None:
        return pd.to_datetime(s, errors="coerce")
    else:
        return pd.to_datetime(s, format=DT_FMT, errors="coerce")

def overlap_hours(a_start, a_end, b_start, b_end) -> float:
    # returns overlap duration in hours (>=0)
    latest_start = max(a_start, b_start)
    earliest_end = min(a_end, b_end)
    if earliest_end <= latest_start:
        return 0.0
    return (earliest_end - latest_start).total_seconds() / 3600.0

# =========================
# preprocess: parse times
# =========================
for d in (df1, df2):
    d["dtst_dt"] = parse_dt(d["dtst"])
    d["dten_dt"] = parse_dt(d["dten"])

# もし NaT が多いと一致判定できないので、ここで確認用（printしたくなければ消してOK）
print("df1 dt parse NaT:", df1["dtst_dt"].isna().sum(), df1["dten_dt"].isna().sum())
print("df2 dt parse NaT:", df2["dtst_dt"].isna().sum(), df2["dten_dt"].isna().sum())

# center time
df1["tcenter"] = df1["dtst_dt"] + (df1["dten_dt"] - df1["dtst_dt"]) / 2
df2["tcenter"] = df2["dtst_dt"] + (df2["dten_dt"] - df2["dtst_dt"]) / 2

# ensure lat/lon numeric (完全一致をやめて距離判定なのでfloatに)
df1["lat"] = pd.to_numeric(df1["lat"], errors="coerce")
df1["lon"] = pd.to_numeric(df1["lon"], errors="coerce")
df2["lat"] = pd.to_numeric(df2["lat"], errors="coerce")
df2["lon"] = pd.to_numeric(df2["lon"], errors="coerce")

# =========================
# main matching (list all candidates)
# =========================
time_margin = pd.Timedelta(hours=TIME_MARGIN_H)

candidates_rows = []   # 全候補（リスト出力用）
best_rows = []         # ベスト一致
notfound_rows = []     # 未一致（df1）

# df2 を時刻で絞るために、dtst/dtenの列だけ先に持っておく（高速化の最低限）
df2_dtst = df2["dtst_dt"].values
df2_dten = df2["dten_dt"].values
df2_tcen = df2["tcenter"].values
df2_lat = df2["lat"].values
df2_lon = df2["lon"].values

for i in range(len(df1)):
    r1 = df1.iloc[i]

    # 欠損があると判定不能なのでスキップ扱い（必要ならログへ）
    if pd.isna(r1["dtst_dt"]) or pd.isna(r1["dten_dt"]) or pd.isna(r1["lat"]) or pd.isna(r1["lon"]):
        notfound_rows.append({
            "df1_i": i,
            "reason": "df1 missing dt/lat/lon",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A: 時間の粗い候補抽出（±12hの範囲に“重なり得る”もの）
    a_start = r1["dtst_dt"] - time_margin
    a_end   = r1["dten_dt"] + time_margin

    # df2イベント区間 [dtst_dt, dten_dt] が [a_start, a_end] と重なり得る条件
    # b_start <= a_end かつ b_end >= a_start
    time_mask = (df2_dtst <= a_end) & (df2_dten >= a_start)

    idx_time = np.where(time_mask)[0]
    if idx_time.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "no time candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- B: 位置（円距離）で絞る
    dlat = df2_lat[idx_time] - float(r1["lat"])
    dlon = df2_lon[idx_time] - float(r1["lon"])
    dist_grid = np.sqrt(dlat*dlat + dlon*dlon)

    pos_mask = dist_grid <= POS_RADIUS_GRID
    idx_tp = idx_time[pos_mask]
    dist_tp = dist_grid[pos_mask]

    if idx_tp.size == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time ok but no position candidates",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
        })
        continue

    # ---- A(強化): 重なり量 + 中心時刻差で絞る
    # 重なりは「元のイベント区間」同士で評価（マージンは候補抽出だけに使う）
    keep = []
    for k, j in enumerate(idx_tp):
        ov_h = overlap_hours(r1["dtst_dt"], r1["dten_dt"], df2.iloc[j]["dtst_dt"], df2.iloc[j]["dten_dt"])
        center_diff_h = abs((r1["tcenter"] - df2_tcen[j]).total_seconds()) / 3600.0

        if (ov_h >= MIN_OVERLAP_H) and (center_diff_h <= MAX_CENTER_DIFF_H):
            keep.append((j, dist_tp[k], ov_h, center_diff_h))

    if len(keep) == 0:
        notfound_rows.append({
            "df1_i": i,
            "reason": "time&pos ok but overlap/center fail",
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": r1["lat"],
            "df1_lon": r1["lon"],
            "df1_hrid": r1["hrid"],
        })
        continue

    # ---- C: スコアで最良一致を選ぶ（ただし候補リストは全部残す）
    # score = dist_grid + ALPHA_H * center_diff_h
    cand_list = []
    for (j, distg, ov_h, cd_h) in keep:
        score = distg + ALPHA_H * cd_h
        cand_list.append((j, distg, ov_h, cd_h, score))
        candidates_rows.append({
            "df1_i": i,
            "df2_i": int(j),
            "df1_hrid": r1["hrid"],
            "df2_hrid": df2.iloc[j]["hrid"],
            "dist_grid": float(distg),
            "overlap_h": float(ov_h),
            "center_diff_h": float(cd_h),
            "score": float(score),

            # ぱっと確認用：最小限のメタも付与（必要に応じて増やしてOK）
            "df1_dtst": r1["dtst"],
            "df1_dten": r1["dten"],
            "df1_lat": float(r1["lat"]),
            "df1_lon": float(r1["lon"]),
            "df2_dtst": df2.iloc[j]["dtst"],
            "df2_dten": df2.iloc[j]["dten"],
            "df2_lat": float(df2.iloc[j]["lat"]),
            "df2_lon": float(df2.iloc[j]["lon"]),
        })

    # best = min score
    cand_list.sort(key=lambda x: x[4])
    best_j, best_distg, best_ov, best_cd, best_score = cand_list[0]

    best_rows.append({
        "df1_i": i,
        "df2_i": int(best_j),
        "df1_hrid": r1["hrid"],
        "df2_hrid": df2.iloc[j]["hrid"],
        "dist_grid": float(best_distg),
        "overlap_h": float(best_ov),
        "center_diff_h": float(best_cd),
        "score": float(best_score),
        "n_candidates": len(cand_list),

        "df1_dtst": r1["dtst"],
        "df1_dten": r1["dten"],
        "df1_lat": float(r1["lat"]),
        "df1_lon": float(r1["lon"]),
        "df2_dtst": df2.iloc[best_j]["dtst"],
        "df2_dten": df2.iloc[best_j]["dten"],
        "df2_lat": float(df2.iloc[best_j]["lat"]),
        "df2_lon": float(df2.iloc[best_j]["lon"]),
    })

# =========================
# outputs
# =========================
cand_df = pd.DataFrame(candidates_rows).sort_values(["df1_i", "score"])
best_df = pd.DataFrame(best_rows).sort_values(["df1_i"])
nf_df   = pd.DataFrame(notfound_rows).sort_values(["df1_i"])

cand_csv = f"match_candidates_{THS}_A+B+C.csv"
best_csv = f"match_best_{THS}_A+B+C.csv"
nf_csv   = f"match_notfound_{THS}_A+B+C.csv"

cand_df.to_csv(cand_csv, index=False)
best_df.to_csv(best_csv, index=False)
nf_df.to_csv(nf_csv, index=False)

print("saved:", cand_csv)
print("saved:", best_csv)
print("saved:", nf_csv)
print("counts: candidates=", len(cand_df), "best=", len(best_df), "notfound=", len(nf_df))
```

```python
#match_candidates_*_A+B+C.csvで一致した事例をもとのTruecsvから引き算、HK2020では抽出されてない事例のリストを作成
```

```python
THS = '090-70'
datapath = '/mnt/jet12/makoto/extract_senjo/condition_2_re/csv_090-70_25'
csvfile = f'{datapath}/total_4-10_2000-2024_ratio2.0_linear-stationary_cases_{THS}_newarea_True.csv'
candcsv = f'/mnt/jet12/makoto/extract_senjo/coincidence/match_candidates_{THS}_A+B+C.csv'
outcsv = f'{datapath}/reject_HK2020cases_{THS}.csv'

# 元のTrueCSV（全部）
df_all = pd.read_csv(csvfile, dtype={'dtst': str, 'dten': str, 'hrid': str})

# 一致候補CSV（df2_hrid を使う）
cand = pd.read_csv(candcsv, dtype={'df2_hrid': str})

# match_candidates に載っている「一致した df2_hrid」をユニークで回収
matched_hrids = (
    cand['df2_hrid']
    .dropna()
    .astype(str)
    .unique()
)

# TrueCSVから一致分を引き算（= HK2020では抽出されてない）
reject = df_all[~df_all['hrid'].isin(matched_hrids)].copy()

reject.to_csv(outcsv, index=False)
print("matched (unique df2_hrid):", len(matched_hrids))
print("total in TrueCSV:", len(df_all))
print("reject (not in HK2020 match):", len(reject))
print("matched pairs (rows):", len(cand))  # 33 になるはず
print("saved:", outcsv)
```
